```python
# 1. Variables & Data Types
name = "Vishnu S"
age = 20
height = 5.9
is_student = True
print(f"Name: {name}, Age: {age}, Height: {height}, Student: {is_student}")

# 2. Operators
a = 10
b = 3
print("Addition:", a + b)
print("Division:", a / b)
print("Is a greater than b?", a > b)

# 3. Loop example
print("\nMultiplication Table of 5:")
for i in range(1, 11):
    print(f"5 x {i} = {5*i}")

# 4. Function example
def calculate_average(numbers):
    return sum(numbers) / len(numbers)

scores = [85, 90, 78, 92, 88]
print("\nAverage score:", calculate_average(scores))

# 5. While loop example
count = 1
while count <= 5:
    print(f"Count: {count}")
    count += 1
```
